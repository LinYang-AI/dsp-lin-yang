```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
```

```python
dataset = np.random.randint(0, 100, size=(10,3))
```

```python
dataset
```

```python
fig, ax = plt.subplots()
ax.scatter(dataset[:,0], dataset[:,1], dataset[:,2], cmap='viridis')
cbar = plt.colorbar(ScalarMappable(cmap='viridis'))
cbar.set_label('Column 3')
ax.set_xlabel('Column 1')
ax.set_ylabel('Column 2')
plt.show()
```
