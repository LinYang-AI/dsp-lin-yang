```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
```

```python
dataset = np.random.randint(0, 100, size=(10, 3))
dataset
```

```python
fig, ax = plt.subplots()
ax.scatter(dataset[:, 0], dataset[:, 1], c=dataset[:, 2], cmap='viridis')
cbar = plt.colorbar(ScalarMappable(cmap='viridis'))
cbar.set_label('Column 3')
cbar.set_label('Column 2')
cbar.set_label('Column 1')
plt.show()
```
